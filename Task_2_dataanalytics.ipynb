```python
import pandas as pd

file_path = '/content/drive/MyDrive/da 2/dataset for assignment 2.csv'
data = pd.read_csv(file_path)
data.head()
```

```python
# Gender distribution
gender_distribution = data['Gender'].value_counts()
print(gender_distribution)

# Age distribution
age_distribution = data['Age'].describe()
print(age_distribution)

# Location distribution
location_distribution = data['Location'].value_counts()
print(location_distribution)

# Activity Level distribution
activity_level_distribution = data['Activity Level'].value_counts()
print(activity_level_distribution)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Age vs. Calories Burned
sns.boxplot(x='Age', y='Calories Burned', data=data)
plt.title('Age vs. Calories Burned')
plt.show()

# Gender vs. Distance Travelled
sns.boxplot(x='Gender', y='Distance Travelled (km)', data=data)
plt.title('Gender vs. Distance Travelled')
plt.show()

# Location vs. App Sessions
sns.boxplot(x='Location', y='App Sessions', data=data)
plt.title('Location vs. App Sessions')
plt.show()

# Activity Level vs. Calories Burned
sns.boxplot(x='Activity Level', y='Calories Burned', data=data)
plt.title('Activity Level vs. Calories Burned')
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Gender distribution
gender_counts = data['Gender'].value_counts()
plt.figure(figsize=(8, 6))
sns.barplot(x=gender_counts.index, y=gender_counts.values, palette='viridis')
plt.title('Gender Distribution')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()
```

```python
# Age distribution
plt.figure(figsize=(10, 6))
sns.histplot(data['Age'], bins=10, kde=True, color='blue')
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```
